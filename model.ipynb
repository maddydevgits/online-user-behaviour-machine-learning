```python
import pandas as pd 
dataset=pd.read_csv('online_classroom_data.csv')
```

```python
dataset.replace(',','.',regex=True,inplace=True)
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
x=dataset.iloc[:,1:10]
```

```python
x
```

```python
y=dataset.iloc[:,-1]
```

```python
y
```

```python
x=x.values
```

```python
x
```

```python
y=y.values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier1=LogisticRegression()
classifier1.fit(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier2=KNeighborsClassifier()
classifier2.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier3=DecisionTreeClassifier()
classifier3.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier4=RandomForestClassifier()
classifier4.fit(x_train,y_train)
```

```python
from sklearn.svm import SVC
classifier5=SVC()
classifier5.fit(x_train,y_train)
```

```python
y_prediction1=classifier1.predict(x_test)
```

```python
y_prediction2=classifier2.predict(x_test)
```

```python
y_prediction3=classifier3.predict(x_test)
```

```python
y_prediction4=classifier4.predict(x_test)
```

```python
y_prediction5=classifier5.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
c1score=accuracy_score(y_test,y_prediction1)
print(c1score)
```

```python
from sklearn.metrics import accuracy_score
c2score=accuracy_score(y_test,y_prediction2)
print(c2score)
```

```python
from sklearn.metrics import accuracy_score
c3score=accuracy_score(y_test,y_prediction3)
print(c3score)
```

```python
from sklearn.metrics import accuracy_score
c4score=accuracy_score(y_test,y_prediction4)
print(c4score)
```

```python
from sklearn.metrics import accuracy_score
c5score=accuracy_score(y_test,y_prediction5)
print(c5score)
```

```python
import matplotlib.pyplot as plt
scores=[c1score,c2score,c3score,c4score,c5score]
models=['Logistic','KNN','DT','RF','SVC']
fig=plt.figure(figsize=(10,5))
plt.bar(models,scores,color='green',width=0.5)
plt.xlabel('MachineLearningModels')
plt.ylabel('AccuracyScores')
plt.show()
```
